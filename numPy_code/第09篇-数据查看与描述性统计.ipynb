## 第一眼：看规模和结构

````python
import pandas as pd
import numpy as np

# 用项目数据演示
df = pd.read_csv('全国城市数据.csv',encoding='gbk')

# 行列数
print(df.shape)       # (343, 7) → 343行，7列

# 列名列表
print(df.columns.tolist())

# 数据类型
print(df.dtypes)
````

## df.info()：一次性看清楚

````python
df.info()
````

## 查看前几行和后几行

````python
df.head()       # 默认前5行
df.head(10)     # 前10行
df.tail(3)      # 后3行
df.sample(5)    # 随机抽5行（看数据多样性）
````

---

## 缺失值检查

````python
# 每列缺失值数量
df.isnull().sum()
````

````python
# 缺失比例（更直观）
(df.isnull().sum() / len(df) * 100).round(2)
````

````python
# 哪些行有缺失值
df[df.isnull().any(axis=1)]
````

---

## 重复值检查

````python
# 重复行数量
df.duplicated().sum()
````

````python
# 查看重复的行
df[df.duplicated()]
````

````python
# 某列是否有重复值（比如 ID 列不应该重复）
df['城市名称'].duplicated().sum()
````

---

## 值分布：value_counts()

````python
# 类别列的频率统计
df['城市级别'].value_counts()
````

````python
# 归一化为比例
df['城市级别'].value_counts(normalize=True).round(3)
````

````python
# 数值列的分布（分箱统计）
df['常住人口(万人)'].value_counts(bins=5)
````

---

## 唯一值

````python
print(df['省份'].nunique())         # 31，有多少个不同省份
print(df['省份'].unique())           # 返回所有唯一值的数组
print(df['城市级别'].value_counts()) # 每个级别的城市数量
````

## 实战：一份陌生数据集的快速体检流程

把上面的内容整合成一个函数，以后拿到新数据集直接调用：

````python
def quick_check(df, name="数据集"):
    """快速查看数据集基本情况"""
    print(f"\n{'='*45}")
    print(f"  📊 {name} 快速体检报告")
    print(f"{'='*45}")

    print(f"\n【规模】{df.shape[0]} 行 × {df.shape[1]} 列")

    print(f"\n【数据类型】")
    for col, dtype in df.dtypes.items():
        print(f"  {col}: {dtype}")

    print(f"\n【缺失值】")
    missing = df.isnull().sum()
    missing = missing[missing > 0]
    if len(missing) == 0:
        print("  无缺失值 ✅")
    else:
        for col, cnt in missing.items():
            print(f"  {col}: {cnt} ({cnt/len(df)*100:.1f}%)")

    print(f"\n【重复行】{df.duplicated().sum()} 行")

    print(f"\n【数值列统计】")
    print(df.describe().round(2))

    print(f"\n【类别列分布（前5）】")
    for col in df.select_dtypes(include=["object", "string"]).columns:
        print(f"\n  {col}:")
        print(df[col].value_counts().head(5).to_string(header=False))

# 使用示例
df = pd.read_csv('全国城市数据.csv', encoding = 'gbk')
quick_check(df, "全国城市数据")
````

这个 `quick_check` 函数我自己一直在用，**送给你了**，直接复制走。

## 本篇小结

| 方法 | 用途 |
|------|------|
| `df.shape` | 行列数 |
| `df.dtypes` | 各列数据类型 |
| `df.info()` | 完整数据概览 |

| `df.sample(n)` | 随机采样查看 |
