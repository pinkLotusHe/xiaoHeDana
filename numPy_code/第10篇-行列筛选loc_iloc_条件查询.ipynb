嗨，我是小荷。

上一篇我们学会了"看数据"，这一篇学"取数据"。

你可能会问：NumPy 那边不是已经学过索引和切片了吗？Pandas 有什么不一样？

有一个核心区别：**NumPy 只有位置索引（数字），Pandas 有标签索引（列名/行名）**。这让 Pandas 的取数方式更灵活，但也更容易搞混。

萧何管账，可以按"第几本"翻账簿（位置），也可以按"哪个郡"找记录（标签）。Pandas 的 loc 和 iloc，就是这两种方式的具体实现。

---

## iloc：按位置取（I = Integer）

`iloc` 永远按**整数位置**取数，用法和 NumPy 二维数组完全一样。

```python
import pandas as pd

df = pd.read_csv('全国城市数据.csv', encoding = 'gbk')

# 取单行（返回 Series）
df.iloc[0]         # 第0行
```

```python
df.iloc[-1]        # 最后一行
```

```python
# 取多行（返回 DataFrame）
df.iloc[0:5]       # 前5行
```

```python
df.iloc[[0, 2, 4]] # 第0、2、4行
```

```python
# 取单个值
df.iloc[0, 1]      # 第0行，第1列
```

```python
# 取行列区间
df.iloc[0:3, 0:3]  # 前3行，前3列
```

```python
df.iloc[:, 2]      # 所有行，第2列
```

---

## loc：按标签取（L = Label）

`loc` 按**行标签**和**列名**取数，更符合"按名字找数据"的直觉。

```python
# 按列名取列（最常用！）
df.loc[:, '城市名称']               # 等价于 df['城市名称']
```

```python
df.loc[:, ['省份', '城市名称', '城市级别']]  # 取多列
```

```python
# 按行索引取行
df.loc[0]          # 行索引为 0 的行
```

```python
df.loc[0:4]        # 注意：loc 的切片是闭区间，包含第4行！
```

```python
# 取单个值
df.loc[0, '城市名称']    # 第0行，城市名称列
```

> ⚠️ **iloc vs loc 的切片区别**：
> - `iloc[0:5]` → 取第 0、1、2、3、4 行（**不含**第5行，像 Python）
> - `loc[0:5]` → 取索引 0、1、2、3、4、5 的行（**含**第5行）
>
> 这个区别坑过无数人，记清楚。

## 条件查询：筛选满足条件的行

这才是 Pandas 筛选数据最常用的方式，逻辑和 NumPy 的布尔索引完全一样。

### 单条件筛选

```python
# 筛选一线城市
first_tier = df[df['城市级别'] == '一线']
print(first_tier[['省份', '城市名称']])

# 筛选人口超过1000万的城市
big_cities = df[df['常住人口(万人)'] > 1000]

# 筛选面积小于5000的城市
small_area = df[df['面积(km2)'] < 5000]
```

### 多条件筛选

```python
# AND：用 & 连接，每个条件要加括号
result = df[(df['城市级别'] == '二线') & (df['常住人口(万人)'] > 500)]
result
```

```python
# OR：用 | 连接
result = df[(df['城市级别'] == '一线') | (df['城市级别'] == '准一线')]
result
```

```python
# NOT：用 ~ 取反
not_other = df[~(df['城市级别'] == '其他')]
not_other
```

### isin()：匹配多个值

```python
# 筛选某几个省份
target_provinces = ['广东省', '浙江省', '江苏省']
south_cities = df[df['省份'].isin(target_provinces)]
south_cities
```

```python
# 筛选非目标城市
other_cities = df[~df['省份'].isin(target_provinces)]
other_cities
```

### between()：区间筛选

```python
# 筛选人口在 100~500 万之间的城市
mid_cities = df[df['常住人口(万人)'].between(100, 500)]
mid_cities
```

### str 系列：字符串匹配

```python
# 城市名包含"市"
has_shi = df[df['城市名称'].str.contains('市')]
has_shi
```

```python
# 省份以"省"结尾
provinces = df[df['省份'].str.endswith('省')]
provinces
```

```python
# 忽略大小写
df[df['城市名称'].str.contains('Beijing', case=False)]
```

---

## query()：写 SQL 风格的查询

对 SQL 熟悉的同学会喜欢这个：

```python
# 等价于 df[(df['城市级别'] == '一线') & (df['常住人口(万人)'] > 500)]
result = df.query("城市级别 == '一线' and `常住人口(万人)` > 500")

# 用变量
min_pop = 500
result = df.query("`常住人口(万人)` > @min_pop")  # @ 引用变量
result
```

## 实战：从全国城市数据里精准提取

```python
import pandas as pd

df = pd.read_csv('全国城市数据.csv', encoding = 'gbk')

# 任务1：找出所有一线和准一线城市，按人口降序排列
top_cities = df[df['城市级别'].isin(['一线', '准一线'])]\
    .sort_values('常住人口(万人)', ascending=False)\
    [['省份', '城市名称', '城市级别', '常住人口(万人)']]\
    .reset_index(drop=True)
print("一线+准一线城市（按人口降序）:")
print(top_cities)

# 任务2：找出面积最大的10个城市
largest = df.nlargest(10, '面积(km2)')[['城市名称', '省份', '面积(km2)']]
print("\n面积最大的10个城市:")
print(largest)

# 任务3：广东省所有城市
guangdong = df[df['省份'] == '广东省'][['城市名称', '城市级别', '常住人口(万人)']]
print(f"\n广东省城市（共{len(guangdong)}个）:")
print(guangdong)
```

## 选列的简洁写法汇总

```python
# 选单列
df['列名']        # 返回 Series
df[['列名']]      # 返回 DataFrame（双括号！）

# 选多列
df[['列1', '列2', '列3']]

# 按类型选列
df.select_dtypes(include='number')    # 所有数值列
df.select_dtypes(include='object')    # 所有字符串列

# 按列名过滤（列名包含某字符串）
df.filter(like='人口')   # 选出列名含"人口"的所有列
df.filter(regex='^火星')  # 列名以"火星"开头的列
```

## 本篇小结

| 方法 | 用途 | 特点 |
|------|------|------|
| `iloc[行, 列]` | 按位置取 | 整数，切片不含末端 |
| `loc[行, 列]` | 按标签取 | 列名/索引，切片含末端 |
| `df[条件]` | 条件筛选 | 布尔索引 |
| `isin([...])` | 多值匹配 | 等价于 SQL IN |
| `between(a, b)` | 区间筛选 | 闭区间 |
| `str.contains()` | 字符串筛选 | 支持正则 |
| `query()` | SQL 风格查询 | 代码更简洁 |
