## Pandas 是什么？

Pandas 是基于 NumPy 的数据分析库，提供两种核心数据结构：

- **Series**：一维带标签的数组（像一列数据）
- **DataFrame**：二维带标签的表格（像一张 Excel 表）

````python
import pandas as pd
import numpy as np
````

---

## Series：一列数据

### 创建 Series

````python
# 从列表创建
scores = pd.Series([85, 92, 78, 96, 88])
print(scores)
# 0    85
# 1    92
# 2    78
# 3    96
# 4    88
# dtype: int64

# 自定义索引
scores = pd.Series(
    [85, 92, 78, 96, 88],
    index=['张三', '李四', '王五', '赵六', '小荷']
)
print(scores['小荷'])  # 88
````

### Series 的核心属性

````python
print(scores.index)    # 索引
print(scores.values)   # 值（NumPy 数组）
print(scores.dtype)    # 数据类型
print(scores.shape)    # 形状，如 (5,)
print(len(scores))     # 长度
````

### Series 的常用操作

````python
# 支持向量化运算（广播机制）
print(scores + 5)       # 所有成绩加5分
print(scores[scores > 90])  # 筛选大于90的

# 统计方法
print(scores.mean())    # 均值
print(scores.max())     # 最大值
print(scores.describe()) # 一键出描述统计
````

> Series 本质上是「有名字的 NumPy 数组」，所有 NumPy 的向量化操作在 Series 上都能用。

## DataFrame：二维数据表

### 创建 DataFrame

````python
# 从字典创建（最常用）
data = {
    '姓名': ['张三', '李四', '王五', '赵六'],
    '语文': [85, 92, 78, 96],
    '数学': [90, 68, 87, 75],
    '英语': [88, 95, 82, 91]
}
df = pd.DataFrame(data)
print(df)
#    姓名  语文  数学  英语
# 0  张三   85   90   88
# 1  李四   92   68   95
# 2  王五   78   87   82
# 3  赵六   96   75   91

# 从 NumPy 数组创建
arr = np.random.randint(60, 100, size=(4, 3))
df2 = pd.DataFrame(arr, columns=['语文', '数学', '英语'])
````

### DataFrame 的核心属性

````python
print(df.shape)        # (4, 4) — 行数和列数
print(df.columns)      # 列名
print(df.index)        # 行索引
print(df.dtypes)       # 每列的数据类型
print(df.values)       # 转为 NumPy 数组
````

---

## DataFrame 的基础操作

### 查看数据

````python
df.head(3)      # 前3行
df.tail(2)      # 后2行
df.info()       # 数据概览（行数、列数、数据类型、内存）
df.describe()   # 数值列的描述统计（均值、标准差、分位数等）
````

`df.info()` 的输出长这样：
```
<class 'pandas.core.frame.DataFrame'>
RangeIndex: 4 entries, 0 to 3
Data columns (total 4 columns):
 #   Column  Non-Null Count  Dtype
---  ------  --------------  -----
 0   姓名      4 non-null      object
 1   语文      4 non-null      int64
 2   数学      4 non-null      int64
 3   英语      4 non-null      int64
dtypes: int64(3), object(1)
memory usage: 260.0+ bytes
```

> 拿到新数据集，先跑 `df.info()` 和 `df.describe()`，这是最快的"数据体检"方式。

### 取列

````python
# 取单列（返回 Series）
df['语文']
df.语文       # 属性访问，列名是合法变量名时可以用

# 取多列（返回 DataFrame）
df[['语文', '数学']]
````

### 取行

````python
# 按索引取行
df.iloc[0]      # 第0行（位置索引）
df.iloc[1:3]    # 第1到第2行

# 按标签取行
df.loc[0]       # 索引为0的行
````

### 新增和修改列

````python
# 新增总分列
df['总分'] = df['语文'] + df['数学'] + df['英语']

# 新增等级列
df['等级'] = df['总分'].apply(lambda x: 'A' if x >= 270 else 'B')

print(df)
#    姓名  语文  数学  英语   总分 等级
# 0  张三   85   90   88  263   B
# 1  李四   92   68   95  255   B
# 2  王五   78   87   82  247   B
# 3  赵六   96   75   91  262   B
````

---

## Series vs DataFrame 的关系

```
DataFrame = 多个 Series 拼在一起

df['语文']  →  返回一个 Series
df[['语文', '数学']]  →  返回一个 DataFrame
```

可以这样理解：**DataFrame 是账本，Series 是账本里的某一列**。萧何的账本上有日期列、物资列、数量列，每一列单独拿出来就是一个 Series。

## 实战：快速了解一份陌生数据集

````python
import pandas as pd
import numpy as np

# 模拟一份电商订单数据
np.random.seed(42)
n = 100
df = pd.DataFrame({
    '订单ID': range(1001, 1001 + n),
    '用户ID': np.random.randint(1, 20, n),
    '商品类别': np.random.choice(['数码', '服装', '食品', '图书'], n),
    '金额': np.random.uniform(10, 1000, n).round(2),
    '数量': np.random.randint(1, 5, n),
    '是否退款': np.random.choice([True, False], n, p=[0.1, 0.9])
})

# === 数据体检五步走 ===
print("【1】基本信息")
print(f"  行数: {df.shape[0]}, 列数: {df.shape[1]}")

print("\n【2】数据类型")
print(df.dtypes)

print("\n【3】前5行")
print(df.head())

print("\n【4】数值统计")
print(df.describe().round(2))

print("\n【5】缺失值检查")
print(df.isnull().sum())
````

这五步，是我每次拿到新数据集必跑的操作，养成习惯，省很多后续的排查时间。

## 本篇小结

| 对象 | 维度 | 特点 | 类比 |
|------|------|------|------|
| Series | 一维 | 带标签的数组 | 账本里的一列 |
| DataFrame | 二维 | 带行列标签的表格 | 完整的账本 |

**记住三件事：**
1. `df.info()` + `df.describe()` 是拿到数据必跑的第一步
2. 取单列用 `df['列名']`（返回 Series），取多列用 `df[['列名1','列名2']]`（返回 DataFrame）
3. Series 和 NumPy 数组很像，NumPy 能做的操作 Series 基本都能做

## 课后练习

````python
import pandas as pd

data = {
    '城市': ['北京', '上海', '广州', '深圳', '杭州'],
    'GDP(亿元)': [41610, 43214, 28839, 32387, 20035],
    '人口(万人)': [2184, 2489, 1882, 1766, 1237],
    '一线城市': [True, True, False, False, False]
}
df = pd.DataFrame(data)

# 任务1：打印 df 的基本信息（行列数、类型）
# 任务2：计算每个城市的人均 GDP（单位：万元/人），新增一列
# 任务3：筛选出 GDP 超过 3 万亿的城市
# 任务4：找出人口最多的城市名称
````

````python
# 任务1：打印 df 的基本信息（行列数、类型）
df.info()
````

````python
# 任务2：计算每个城市的人均 GDP（单位：万元/人），新增一列
df['人均GDP(万元/人)'] = df['GDP(亿元)']/df['人口(万人)']
df
````

````python
# 任务3：筛选出 GDP 超过 3 万亿的城市
df[df['GDP(亿元)'] > 30000]
````

````python
# 任务4：找出人口最多的城市名称
max_index = df['人口(万人)'].idxmax()
df.loc[max_index, '城市'] 
````
