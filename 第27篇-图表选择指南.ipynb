## 一、选图前先问四个问题

**1. 我的数据是什么类型？**

| 数据类型 | 例子 |
|---------|------|
| 分类数据 | 城市、品类、会员等级 |
| 连续数据 | 金额、年龄、温度 |
| 时间数据 | 日期、月份、季度 |
| 比例数据 | 占比、百分比 |

**2. 我想表达什么？**

- 对比（谁大谁小）
- 趋势（随时间变化）
- 分布（集中在哪里，有无异常）
- 关系（两变量有没有关联）
- 构成（整体中各部分占多少）

**3. 有几组数据？**

- 1 组：看分布
- 2 组：对比
- 多组：多维比较

**4. 给谁看？**

- 技术人员：可以用复杂图（如小提琴图、散点矩阵）
- 管理层/非技术：用简单直观的（柱状图、折线图、饼图）

---

## 二、按表达目的选图

### 2.1 对比

**柱状图** — 分类数据间的对比，最通用

- 纵向柱状图：类别 ≤8，名字短
- 水平柱状图：类别名字长，或类别较多（8~20个）
- 分组柱状图：比较多个时间点/条件下的多组数据
- 堆叠柱状图：既对比总量，又看各组成部分


```python
类别少、名短 → 纵向柱状图
类别名字长   → 水平柱状图
两个时期对比 → 分组柱状图
看构成+对比  → 堆叠柱状图
```

**雷达图** — 多维度打分对比（如商品综合评分）

> ⚠️ 注意：雷达图维度超过 7 个就很难看清楚了。

---

### 2.2 趋势

**折线图** — 时间序列，连续变化

- 单线：一个指标的时间趋势
- 多线：多指标或多组的趋势对比


```python
时间序列 → 折线图（首选）
时间+量感 → 面积图
多条线太密 → 改用分面（FacetGrid）
```

> ⚠️ **误区**：用折线图连接没有顺序关系的分类数据（如"城市"），这是错的！折线暗示"相邻类别之间有过渡"，城市之间没有这种关系，应该用柱状图。

---

### 2.3 分布

| 图表 | 适合 | 特点 |
|------|------|------|
| 直方图 | 单变量连续分布 | 看形态、区间频率 |
| KDE | 单/多变量分布对比 | 平滑，适合多组叠加 |
| 箱线图 | 多组分布对比 | 中位数、四分位、异常值 |
| 小提琴图 | 多组分布对比 | 分布形态更精细 |


```python
单组探索分布    → 直方图 + KDE
多组分布对比    → 箱线图（简洁）或小提琴图（详细）
需要看原始数据  → 条带图（stripplot）叠加在箱线图上
```

---

### 2.4 关系

| 图表 | 适合 |
|------|------|
| 散点图 | 两连续变量关系 |
| 气泡图 | 两连续变量 + 第三变量 |
| 热力图 | 多变量相关矩阵 |
| 散点矩阵 | 多变量两两关系 |


```python
两变量关系  → 散点图 + 趋势线
三变量      → 气泡图（大小编码第三变量）
多变量全局  → 热力图（相关矩阵）
```

> ⚠️ 相关 ≠ 因果！散点图只能发现"可能有关系"，不能证明因果。

---

### 2.5 构成/占比

**饼图** — 整体的部分构成

- 使用条件：分类 ≤6 个，且各类占比差异明显
- 环形饼图（Donut）：中间可以放核心数字，更美观

**堆叠柱状图** — 对比多个时间点的构成变化

**面积图** — 多个系列的趋势 + 量感


```python
简单占比（类别少） → 饼图
占比随时间变化    → 堆叠面积图
多组构成对比      → 堆叠柱状图
```

> ⚠️ **饼图误区**：
> - 类别太多（>6）→ 改用水平柱状图
> - 各类占比相近 → 人眼很难区分，改用柱状图
> - 3D 饼图 → 会产生视觉扭曲，不推荐

## 三、常见可视化误区

### 误区1：Y 轴不从 0 开始（夸大差异）

夸大差异，来看具体怎么操作：


```python
❌ 错误：Y轴从 95 开始，让微小差异看起来很大
✅ 正确：柱状图 Y 轴必须从 0 开始
   （折线图可以不从0，但要标明轴范围）
```

### 误区2：用 3D 图（难以读数）

难以读数，来看具体怎么操作：


```python
❌ 3D 柱状图、3D 饼图 → 透视变形，读不准数值
✅ 改用 2D 图 + 颜色区分
```

### 误区3：颜色太多（眼花缭乱）

颜色是图表的第一印象，来看看怎么设置：


```python
❌ 一张图用了 10 种颜色
✅ 最多 5-6 种颜色；强调 1-2 个，其余用灰色
```

### 误区4：没有标题和轴标签

来看看误区4：没有标题和轴标签怎么实现：


```python
❌ 图表没有标题，看不懂在说什么
✅ 每张图：主标题 + X轴标签 + Y轴标签 + 必要的数值标注
```

### 误区5：信息太密

来看看误区5：信息太密怎么实现：


```python
❌ 一张图放 20 条折线，全部颜色不同
✅ 超过 5-6 条线时，用 FacetGrid 分图展示，或只保留关键几条
```

### 误区6：折线连接无序分类

来看看误区6：折线连接无序分类怎么实现：


```python
❌ X轴是"城市"，却用折线图
✅ 城市是无序分类，用柱状图
```

## 四、一张图：选图决策树

来用代码实现一棵决策树：


```python
我的数据 →
├── 随时间变化？
│     ├── 是 → 折线图（单/多线）、面积图
│     └── 否 ↓
│
├── 想看占比/构成？
│     ├── 类别≤6 → 饼图
│     └── 类别>6 / 随时间变化 → 堆叠柱/面积图
│
├── 想看对比（哪个大）？
│     ├── 类别名字短 → 纵向柱状图
│     └── 类别名字长 → 水平柱状图
│
├── 想看分布（数据长啥样）？
│     ├── 单组 → 直方图 + KDE
│     └── 多组 → 箱线图 / 小提琴图
│
└── 想看关系（两变量是否相关）？
      ├── 两个变量 → 散点图 + 回归线
      ├── 三个变量 → 气泡图
      └── 多个变量 → 相关热力图
```

## 五、📝 小结

选图不是感觉，是逻辑：
1. **先想清楚想表达什么**（对比/趋势/分布/关系/构成）
2. **再看数据类型**（分类/连续/时间/比例）
3. **最后考虑受众**（技术 vs 非技术）

掌握了这个框架，面对任何数据，你都能在 30 秒内选出合适的图。

## 六、🏋️ 课后练习

1. 有一张"2024年每月各品类销售额"的数据表，你会用什么图？说明理由。
2. 有一组"1000个用户的年龄"数据，你会用什么图探索分布？
3. 有"广告投入"和"用户数"两个变量，你想看它们有没有关系，用什么图？

练习1：2024年每月各品类销售额，用什么图？
推荐：堆叠面积图 或 多线折线图

理由分析：

数据特征：时间序列（12个月） × 多分类（各品类），属于「趋势 + 构成」双重需求
如果想看各品类的变化趋势 + 总量 → 堆叠面积图：既能看每月总 GMV 的走势，又能看每个品类在其中的占比变化
如果只想对比各品类趋势 → 多线折线图：每条线代表一个品类，走势清晰，但如果品类超过 5~6 个就会太密
如果品类太多（>6） → 折线图 + FacetGrid 分面，或者只保留 TOP5 品类，其余合并为"其他"

⚠️ 误区提醒：不要用饼图！12 个月 × N 个品类，饼图完全没法展示时间维度。

练习2：1000个用户的年龄，用什么图探索分布？
推荐：直方图 + KDE

理由分析：

数据特征：单个连续变量，目标看分布（集中在哪里、有没有异常）
直方图：最直观，能看出年龄段区间的人数分布，比如 20~30 岁最多
KDE（核密度估计）：叠加在直方图上，看平滑的分布形态，有无双峰、偏斜
如果还想看具体统计量 → 叠加一条均值/中位数竖线

补充方案：

箱线图：快速看中位数、四分位、异常值，但丢失了分布形态细节
小提琴图：箱线图 + KDE 的结合，但只有一组数据时优势不大
💡 实操建议：先用 df['年龄'].describe() 看基本统计量，再画直方图确认形态。

练习3：广告投入和用户数，想看有没有关系，用什么图？
推荐：散点图 + 回归线

理由分析：

数据特征：两个连续变量，目标是看「关系」（有没有关联）
散点图：最直接，一个点代表一组数据，X=广告投入，Y=用户数，一眼看趋势方向
叠加回归线（sns.regplot 或 px.scatter(trendline='ols')）：量化趋势方向，R² 值说明相关性强度
如果想更精确 → 再算一下 Pearson 相关系数，配合相关热力图

总结：
|题目	|核心问题|	推荐图表|	一句话理由|
|-----|-----|-----|-----|
|每月各品类销售额|	趋势 + 构成|	堆叠面积图 / 多线折线图	|时间序列多分类，要看走势和占比|
|1000个用户年龄|	分布|	直方图 + KDE|	单连续变量，看集中和异常|
|广告投入 vs 用户数|	关系|	散点图 + 回归线|	两连续变量，看有没有关联|
